```python
print("Even numbers from 2 to 20:")

for num in range(2, 21):
    if num % 2 == 0:
        print(num)
```

```python
names = ["Ali", "Sara", "Hassan", "Zara", "Usman"]

print("Names in uppercase:")
for name in names:
    print(name.upper())
```

```python
print("Numbers from 10 to 1:")

for num in range(10, 0, -1):
    print(num)
```

```python
numbers = [4, 9, 12, 15, 22, 30, 41]

print("Numbers divisible by 3:")
for n in numbers:
    if n % 3 == 0:
        print(n)
```

```python
print("Squares from 1 to 10:")

for i in range(1, 11):
    print(i, "squared =", i**2)
```

```python
temps_celsius = [0, 10, 22, 30, 35]

print("Temperature conversion:")
for c in temps_celsius:
    f = (c * 9/5) + 32
    print(c, "C =", f, "F")
```

```python
numbers = [8, 12, 15, 20, 5]

total = 0

for num in numbers:
    total += num

print("Total sum =", total)
```

```python
word = "University"

print("Characters in the string:")
for char in word:
    print(char)
```

```python
words = ["laptop", "pen", "notebook", "desk", "calculator", "book"]

print("Words with more than 5 letters:")
for w in words:
    if len(w) > 5:
        print(w)
```
