```python
s={"Ali":[80,90,70],"Sara":[85,95,75],"Ahmed":[60,70,65]}

print({k:sum(v)/3 for k,v in s.items()})

sub=[v[1] for v in s.values()]
print(max(sub),min(sub))

print(sorted(s,key=lambda x:sum(s[x])/3,reverse=True))
```

```python
products = {"Apple":[5,10], "Mango":[10,5]}
cart = {}
cart["Apple"] = 2
cart["Mango"] = 1
total = sum(products[p][0]*q for p,q in cart.items())

total = total - total*0.10 + total*0.05

print(total)
```
